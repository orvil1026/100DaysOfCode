```python
import random
logo = """
.------.            _     _            _    _            _    
|A_  _ |.          | |   | |          | |  (_)          | |   
|( \/ ).-----.     | |__ | | __ _  ___| | ___  __ _  ___| | __
| \  /|K /\  |     | '_ \| |/ _` |/ __| |/ / |/ _` |/ __| |/ /
|  \/ | /  \ |     | |_) | | (_| | (__|   <| | (_| | (__|   < 
`-----| \  / |     |_.__/|_|\__,_|\___|_|\_\ |\__,_|\___|_|\_\\
      |  \/ K|                            _/ |                
      `------'                           |__/           
"""
cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
play_again=True

def deal_card():
    
    return random.choice(cards)

def create_deck(deck):
    for i in range(2):
        deck.append(deal_card())
    



def calculate_score(cards):
    
  
    if sum(cards)==21:
        return 0
    elif sum(cards) >21:
        if 11 in deck:
            deck.remove(11)
            deck.append(1)
            return sum(cards)
    else:
        return sum(cards)

def compare(user_score,computer_score):
    if user_score==computer_score:
        print("It is an draw!!!")
    elif user_score==0:
        print("User has WON!!!")
    elif computer_score==0:
        print("Computer has WON!!!")
    elif computer_score>21:
        print("Comuter busted!!!\n User WON!!!")
    elif user_score>21:
        print("User busted!!!\n Computer WON!!!")
    elif user_score>computer_score:
        print("User WON!!!!")
    else:
        print("Computer WON!!!")
        

    
user_cards = [ ]
computer_cards = [ ]
game_over=False
continue_draw=True
create_deck(user_cards)
create_deck(computer_cards)


while not game_over:    
    user_score=calculate_score(user_cards)
    computer_score=calculate_score(computer_cards)

    print(f"User cards:{user_cards} || User Score:{user_score}")
    print(f"Computer cards:{computer_cards[0]}")

    if user_score==0 or computer_score==0 or user_score>21:
        game_over=True
    else:

        draw_card=input("Do you want to draw a card? y or n")
        if draw_card=='y':
            user_cards.append(deal_card())
            user_score=calculate_score(user_cards)
            print(f"User cards:{user_cards} || User Score:{user_score}")

        elif draw_card=='n':
            game_over=True
            
    while computer_score!=0 and computer_score<17:
        computer_cards.append(deal_card())
        computer_score=calculate_score(computer_cards)
        
    compare(user_score,computer_score)
```
